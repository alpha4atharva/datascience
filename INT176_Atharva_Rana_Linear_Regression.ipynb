```python
import pandas as pd
```

```python
data = {
    "x":[4,9,8,8,7,4],
    "Y":[52,51,50,52,70,60]
}
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
def split(df:pd.DataFrame,train_size):
    train_df = df.sample(frac=train_size/100)
    test_df = df.drop(train_df.index)
    return train_df,test_df

train_df,test_df = split(df,75)
print(train_df)
print(test_df)
```

```python
def linear_reg(train_df,test_df):
    y_avg = train_df["Y"].mean()
    x_avg = train_df["x"].mean()

    numerator=0
    denominator=0
    for index,row in train_df.iterrows():
        y=row['Y']
        x=row['x']
        numerator+=(y-y_avg)*(x-x_avg)
        denominator+=(y-y_avg)*(y-y_avg)
    m=numerator/denominator
    c=y_avg-m*x_avg
    print("x"," ","actual"," ","predicted")
    for index,row in test_df.iterrows():
        y = row["Y"]
        x = row["x"]
        y_predicted = m*x+c
        print(x," ",y,"     ",y_predicted)
```

```python
linear_reg(train_df,test_df)
```
