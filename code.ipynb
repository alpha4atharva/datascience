```python
import pandas as pd
```

```python
df = pd.read_csv("11_Baseketball_Team_Performance_Analysis.csv")
display(df)
```

```python
# Check for null values
print("Null values in each column:")
print(df.isnull().sum())
print("\nTotal null values:", df.isnull().sum().sum())
```

```python
# Method 1: Drop rows with null values
df_dropped = df.dropna()
print(f"Original rows: {len(df)}")
print(f"After dropping nulls: {len(df_dropped)}")
print(f"Rows removed: {len(df) - len(df_dropped)}")
```
