```python
import pandas as pd
import numpy as np
```

```python
data = {
    "x1":[4,9,8,8,7,4],
    "x2":[9,8,7,6,5,4],
    "x3":[1,2,3,4,5,6],
    "y":[52,51,50,52,70,60]
}
df = pd.DataFrame(data)
df.head()
```

```python
def split(df:pd.DataFrame,train_size):
    train_df = df.sample(frac=train_size/100)
    test_df = df.drop(train_df.index)
    return train_df,test_df

train_df,test_df = split(df,75)
print(train_df)
print(test_df)
```

```python
X_train = train_df[['x1', 'x2', 'x3']].copy()
X_train.insert(0, 'intercept', 1)
y_train = train_df['y']

X_test = test_df[['x1', 'x2', 'x3']].copy()
X_test.insert(0, 'intercept', 1)
y_test = test_df['y']

# Convert to numpy arrays
X_train = X_train.values
y_train = y_train.values
X_test = X_test.values
y_test = y_test.values

print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
```

```python
beta = np.linalg.inv(X_train.T.dot(X_train)).dot(X_train.T).dot(y_train)
print("Beta coefficients (intercept, x1, x2, x3):")
print(beta)
```

```python
# Next cell: Predictions
y_pred_train = X_train.dot(beta)
y_pred_test = X_test.dot(beta)

print("Train predictions:", y_pred_train)
print("Test predictions:", y_pred_test)
print("Actual test y:", y_test)
```
